```python
def preprocess(text):
  doc=nlp(text)
  no_stop_word=[token.text for token in doc if not token.is_stop]
  return (no_stop_word)
```

```python
preprocess("im going to play cricket with friends")
```

```python
def preprocess(sent):
  no_stop_word=[]
  doc=nlp(sent)
  for token in doc:
    if not token.is_stop:
      no_stop_word.append(token.text)
  return no_stop_word
```

```python
preprocess("im going to play cricket with frineds")
```

```python
def pre(text):
  doc=nlp(text)
  not_stp=[]
  for token in doc:
    if not token.is_stop:
      not_stp.append(token.text)
  return " ".join( not_stp)
```

```python
pre("im going to play cricket with friends")
```
